```python
def prepare_data(l, k):
    n = len(l)
    # ll = [l[i:i+k][1] for i in range(n-k-1)]
    ll = [[l[i+j][1] for j in range(k)] for i in range(len(l)-k+1)]
    for i in range(len(ll)):
        ll[i].append(l[i+k-1][0])
    #    ll[i].reverse()
    return ll

def handle_file(filename, nk):
    l = []
    with open(filename, 'r') as f:
        l = [[float(num) for num in line.split('\t')] for line in f]
    l = [l[i] for i in range(len(l)) if l[i][0]>=0.5] # 0.75 human response time is 0.7-3 seconds

    ll = [[0,0] for i in range(len(l))]
    i = 0
    k = 0
    while i < len(l):
        key_t  = l[i][0]
        sum_v = l[i][1]
        j = i + 1
        while j<len(l):
            if l[j][0] - key_t <= 0.045:
                sum_v += l[j][1]
                j = j + 1 
            else:
                break;
        ll[k][1] = sum_v / (j-i)
        ll[k][0] = key_t; 
        # print('key_t = ', key_t, 'sum_v = ', sum_v, ' i =  ', i, ' j = ', j, ' k = ', k)
        # print(*ll, sep = ',')
        i = j; 
        k = k + 1
    ll = ll[:k]
    #print('k =', k )
    # return ll
    return prepare_data(ll, nk)

```

```python
s_label = ['no', 'yes']
s_types = ['granite','carpet','rubber', 'cement','brick']
s_subject = ['1','2','3','4','5','6','7']
s_phi = ['20', '30', '40', '50']

k = 5

all_data = []
for label_ in s_label:
    lt = []
    print('handle label: ', label_)
    for type_i in range(len(s_types)):
        type_ = s_types[type_i]
    # for type_ in s_types:
        ll = []
        print('handle type = ', type_)
        for subject_ in s_subject:
            for phi_ in s_phi:
                filename = "./data/" + label_ + "/" + type_ + "/subject_" + subject_+"/phi_"+phi_+".log"
                print('handle filaname = ', filename)
                ll.extend(handle_file(filename, k)) # list[TIME, DIFF_J, DIFF_J-1, DIFF_J-2, ..., DIFF_J-K+1]
        # add type 
        for i in range(len(ll)):
            # ll[i].append(type_)
            ll[i].append(type_i+1)
        lt.extend(ll)
    for i in range(len(lt)):
        lt[i].append(label_)
        lt[i].reverse()
    all_data.extend(lt)

### save to csv 
df = pd.DataFrame(data=all_data,columns=['Label', 'Type', 'Time', 'DF1', 'DF2', 'DF3', 'DF4', 'DF5']) 
df.head()
df.to_csv('robocane.csv', index=False, header=False)
print('done!')
```
